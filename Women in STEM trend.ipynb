# Women in STEM Time series sub-plots

## Trend of enrollment of Female Student in Higher Studies like Engineering, Science and Technology and Medicine from year 2012-2017, Nepal 

```python
#import necessary libraries
import pandas as pd
from datetime import datetime, timedelta
from matplotlib import pyplot as plt
from matplotlib import dates as mpl_dates
%matplotlib inline
```

```python
#List of Dates
dates = [datetime (2012, 1, 1),datetime (2013, 1, 1), datetime (2014, 1, 1), datetime (2015, 1, 1),datetime (2016, 1, 1),datetime (2017, 1, 1)]

#List of variables
BE= [501,2077,1660,110,3336,2322]
MBBS= [1003,7204,6333,7389,11516,11335]
Bsc = [2510,2105,5853,6259, 8085,11095]

ME= [41,0,75,20,53,60]
MD= [535,90,324,444,531,470]
MSc= [444,0,159,545,290,1898]

PhdE= [4,0,1,1,1,1]
phM= [0,0,0,0,0,0]
PS= [12,0,5,4,20,33]

plt.style.use('seaborn')
plt.figure(figsize=(14,10))

#Plotting the data
plt.plot_date(dates, BE,'#F6793A', linestyle = 'solid',marker= 'o', label='Bachelors in Engineering') #plot the line
plt.plot (dates, MBBS, '#A95AA1',linestyle = '-',label='MBBS')
plt.plot (dates,Bsc, '#85C0F9',linestyle = 'dashdot',label='Bachelors in Science and Technology')
plt.plot (dates,ME,'#dfc27d', label='Masters in Engineering')
plt.plot (dates,MD,'#a6611a',linestyle = ':', label='Masters in Medicine')
plt.plot (dates,MSc, '#0F2080',label='Masters in Science and Technology')
plt.plot (dates,PhdE, '#f5f5f5',label='PhD in Engineering')
plt.plot (dates,phM, '#80cdc1', linestyle = '--',label='PhD. in Medicine')
plt.plot (dates,PS, '#018571',linestyle = '--',marker= 'o', label='PhD in Science and Technology') 
plt.legend(loc = 'upper left')
plt.grid()
```

Individual graphs 

```python

plt.style.use('seaborn')
plt.figure(figsize=(15,20))

#Using sub-plots with 3 rows and a column
#Sharing the X-axis
fig,(ax1,ax2,ax3) = plt.subplots(nrows=3,ncols=1, sharex=True)
fig.suptitle('Statistics of Women in STEM in higher education(Masters and PhD) from year 2012-2017')#title of the main plot


ax1.plot (dates,ME,'m',linestyle = 'solid', label='Masters in Engineering')
ax2.plot (dates,MD,'#444444',linestyle = ':',marker = '*', label='Masters in Medicine')
ax3.plot (dates,MSc, '#444444',label='Masters in Science and Technology')
ax1.plot (dates,PhdE, 'y',linestyle = '--',label='PhD in Engineering')
ax2.plot (dates,phM, 'c', label='PhD. in Medicine')
ax3.plot (dates,PS, 'y',linestyle = ':',marker= 'o', label='PhD in Science and Technology') 
ax1.legend(loc = 'upper left')
ax2.legend(loc = 'upper left')
ax3.legend(loc = 'upper left')

plt.tight_layout()
plt.show()
```

Bachelors Degree

```python
fig,(ax1,ax2,ax3) = plt.subplots(nrows=3,ncols=1,sharex=True)

plt.style.use('seaborn')
plt.figure(figsize=(15,15))

ax1.plot(dates, BE,'r',linestyle = '--', marker= 'o', label='Bachelors in Engineering') #plot the line
ax2.plot (dates, MBBS, 'k',label='MBBS')
ax3.plot (dates,Bsc, 'b',label='Bachelors in Science and Technology')

#title of the main plot
fig.suptitle('Statistics of Women in STEM (Bachelors) from year 2012-2017')
#location of legend
ax1.legend(loc = 'upper left')
ax2.legend(loc = 'upper left')
ax3.legend(loc = 'upper left')

```

```python
fig,(axs) = plt.subplots(3,2,figsize=(25,15),sharex=True)

plt.style.use('seaborn')

axs[0,1].plot(dates, BE,'r',linestyle = '--', marker= 'o', label='Bachelors in Engineering') #plot the line
axs[1,1].plot (dates, MBBS, 'k',label='MBBS')
axs[2,1].plot (dates,Bsc, 'b',label='Bachelors in Science and Technology')
axs[0,0].plot (dates,ME,'m',linestyle = '--', label='Masters in Engineering')
axs[1,0].plot (dates,MD,'#444444',linestyle = '--',marker = '*', label='Masters in Medicine')
axs[2,0].plot (dates,MSc, '#444444',linestyle = '--',label='Masters in Science and Technology')
axs[0,0].plot (dates,PhdE, 'y',linestyle = '-',label='PhD in Engineering')
axs[1,0].plot (dates,phM, 'c', label='PhD. in Medicine')
axs[2,0].plot (dates,PS, 'y',marker= 'o', label='PhD in Science and Technology') 

axs[0,1].legend(loc = 'upper left')
axs[1,1].legend(loc = 'upper left')
axs[2,1].legend(loc = 'upper left')
axs[0,0].legend(loc = 'upper left')
axs[1,0].legend(loc = 'upper left')
axs[2,0].legend(loc = 'upper left')
plt.figure(figsize=(15,15))

fig.suptitle('Statistics of Women in STEM from year 2012-2017')

plt.tight_layout()
plt.show()
```
