Plotting One dimensional Random walk

```python
#importing necessary libraries
import random
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
X=[]
N=[]
x=0
d=1
n=1000
for i in range (n):
    x = x+d*random.choice([-1,1])
    N.append(i)
    X.append(x)

#Plotting the lists
plt.plot(X)
plt.xlabel('Dispacement')
plt.ylabel('Time')
plt.grid()
plt.title("1D random walk")
plt.grid()
```

```python
#Using function
def walk1D(x,n,d):
    X=[]
    N=[]
    x=0
    for i in range (n):
        x = x+d*random.choice([-1,1])
        N.append(i)
        X.append(x)
        
    return N,X
```

```python

for x in range (5):
    N1,X1 = walk1D(x, n=1000, d=1)
    plt.plot(N1,X1, label = x, color= C[x])
plt.legend()
```

```python
#list of colors for plotting
C=['#765F2E','#A07D2C','#EBE6DB','#57534C','#016c59','#ffffd4','#fed98e','#fe9929','#d95f0e','#993404']
L=['--','-',':','dashdot','solid','dotted','-',':','dashdot','solid']

plt.figure(figsize=(20,10))
for x in range (10):
    N1,X1 = walk1D(x, n=1000, d=1)
    plt.plot(N1,X1, label = x, color= C[x], linestyle= L[x])
   
plt.legend(title="No of loop",loc='upper right')
```
